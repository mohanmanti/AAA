
**Topological Sorting using indegree**

```python
from collections import deque
#function to return list containing vertices in topological order
def topological_sort(adj,V):
  #vector to store indegree of each vertex
  indegree=[0]*V
  for i in range(V):
    for vertex in adj[i]:
      indegree[vertex]+=1
  #queue containing all vertices with 0 indegree
  q=deque()
  for i in range(V):
    if indegree[i]==0:
      q.append(i)
  result=[]
  while q:
    node=q.popleft()
    result.append(node)
    #decrement indegree of all the adjacent vertices
    for neighbor in adj[node]:
      indegree[neighbor]-=1
      #if indegree becomes 0 push it to the queue
      if indegree[neighbor]==0:
        q.append(neighbor)
  #check for cycle
  if len(result)!=V:
    print("Graph contains cycle")
    return []
  return result

n=6 #number of nodes
#edges
edges=[[0,1],[1,2],[2,3],[4,5],[5,2],[5,1]]
#graph represented as adjacency list
adj=[[] for _ in range(n)]
#constructing adjacency list
for i in edges:
  adj[i[0]].append(i[1])
print("Topological sorting of the Graph" ,end=' ')
#performing topological sort
result=topological_sort(adj,n)
#displaying result
for vertex in result:
  print(vertex,end=' ')
```
