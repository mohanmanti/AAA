Breadth First Search

```python
def BFS(graph,V,src):
  visited=[False]*V #mark all vertices as not visited
  queue=[] #create a queue for bfs
  path=[] #create BFS path
  queue.append(src) #push the current source into queue
  visited[src]=True #mark the source node as visited
  while queue:
    u=queue.pop(0) #remove the vertex from queue
    path.append(u) #add the removed vertex to the path
    # get all adjacent vertices of popped vertex u and
    # if a adjacent has not been visited , mark them as visited
    # put it to the queue
    for v in graph[u]:
      if visited[v]==False:
        queue.append(v)
        visited[v]==True
  return path
graph={0:[1,2],1:[3],2:[4,5],3:[],4:[],5:[]}
V=6
src=0
print("BFS traversal order is" , BFS(graph,V,src))
```

Depth First Search


```python
def DFS(graph,V,src):
  stack=[] #create a stack for DFS
  visited=[False]*V #mark all the vertices as not visited
  path=[] #create a path for DFS
  stack.append(src) # push the source vertex into the stack
  while stack:
    #remove the top vertex from the stack and if it is not visited,
    #mark it as visited and add it to the path
    u=stack.pop()
    if visited[u]==False:
      visited[u]=True
      path.append(u)
     # get all adjacent vertices of popped vertex u and
    # if a adjacent has not been visited, push it into the stack
      for v in graph[u]:
        if visited[v]== False:
          stack.append(v)
  return path
#graph={0:[1,2],1:[3],2:[4,5],3:[],4:[],5:[]}
graph={0:[1],1:[2],2:[0,3],3:[4],4:[5,7],5:[6],6:[4,7],7:[]}
V=8
src=0
print("DFS traversal order is ", DFS(graph,V,src))
```
