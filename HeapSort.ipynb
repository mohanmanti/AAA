**Heap Sort**

```python
def heap_sort(A,n):
  #building max-heap
  for i in range(n//2-1,-1,-1):
    heapify(A,i,n)
  for i in range(n-1,-1,-1):
    A[0],A[i]=A[i],A[0] #extracting max
    heapify(A,0,i) #heapify
  return A
def heapify(A,i,n):
  largest=i #current largest element(root)
  l=2*i+1 #left child index
  r=2*i+2 #right child index
  if l<n and A[l]>A[largest]:
    largest=l
  if r<n and A[r]>A[largest]:
    largest=r
  if largest!=i:
    A[i],A[largest]=A[largest],A[i] #swapping heap_root with last element
    heapify(A,largest,n)

A=[4, 10, 3, 5, 1]
print(heap_sort(A,len(A)))
```
