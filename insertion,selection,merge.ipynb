**Insertion Sort**

```python
def insert_sort(A):
  for i in range(1,len(A)):
    k=A[i]
    j=i-1
    while j>=0 and A[j]>k:
        A[j+1]=A[j]
        j=j-1
    A[j+1]=k
  return A
A=[23,1,10,6,2]
print(insert_sort(A))
```

**Selection Sort**

```python
def selec_sort(A,low):
  for i in range(low,len(A)-1):
    min=i
    for j in range(i+1,len(A)):
      if A[j]<A[min]:
        min=j
    A[i],A[min]=A[min],A[i]
  return A
A=[9,15,3,4,2,5]
print(selec_sort(A,0))
```

**Merge Sort**

```python
def merge(left,right,A):
  ll=len(left) #length of left array
  rl=len(right) #length of right array
  i=0 # initializing index for left array
  j=0 #initializing index for right array
  k=0 #initializing index for sorted array
  #comparing the elements of left array and right array
  #and assigining the lower value to the sorted array
  while i<ll and j<rl:
    if left[i]< right[j]:
      A[k]=left[i]
      i+=1
    else:
      A[k]=right[j]
      j+=1
    k+=1
  #adding the left over elements of left array
  #to sorted array
  while i<ll:
    A[k]=left[i]
    i+=1
    k+=1
  #adding the left over elements of right array
  # to sorted array
  while j<rl:
    A[k]=right[j]
    j+=1
    k+=1
  return A

def merge_sort(A):
  n=len(A)
  if n<2:
    return
  mid=n//2
  leftarr=[0]*mid #creating left array
  rightarr=[0]*(n-mid) #creating right array
  for i in range(0,mid):
    leftarr[i]=A[i]
  for i in range(mid,n):
    rightarr[i-mid]=A[i]
  merge_sort(leftarr)
  merge_sort(rightarr)
  merge(leftarr,rightarr,A)
  return A

A=[2,4,1,6,8,5,3,7]
print("sorted array",merge_sort(A))
```
